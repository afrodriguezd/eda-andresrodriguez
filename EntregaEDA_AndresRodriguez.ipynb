```python
print(f"Skewness: {df['Linea110'].skew()}")
print(f"Kurtosis: {df['Linea110'].kurt()}")
```

Skewness (Asimetría): 5.459687059451494

Un valor positivo indica que la distribución está sesgada hacia la derecha.
Este valor es bastante alto, lo que sugiere una asimetría muy pronunciada.
Significa que hay más valores bajos, pero algunos valores extremadamente altos están "estirando" la distribución hacia la derecha.
En términos de ventas, esto podría indicar que la mayoría de las tiendas/períodos tienen ventas relativamente bajas, pero hay algunos casos de ventas excepcionalmente altas.


Kurtosis (Curtosis): 49.33335113536667

Un valor tan alto indica una distribución leptocúrtica (más puntiaguda que una distribución normal).
Sugiere la presencia de valores extremos o atípicos (outliers).
En su contexto, podría significar que hay algunas instancias de ventas muy altas o muy bajas que se alejan significativamente del promedio.

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX

df_dum_bool['PERIODO'] = pd.to_datetime(df_dum_bool['PERIODO'])

df_tienda = df_dum_bool[df_dum_bool['TIENDA'] == 3501]

serie = df_tienda.set_index('PERIODO')['Linea110']

exogenas = df_tienda[['TIPO_TIENDA_Linea','TIPO_TIENDA_Outlet',	'CLIMA_Cálido',	'CLIMA_Frío',	'CANAL_Retail']]

exogenas = exogenas.set_index(serie.index)

exogenas = exogenas.astype(int)

model_sarimax = SARIMAX(serie, exog=exogenas, order=(2, 1, 1), seasonal_order=(1, 1, 1, 12))
model_sarimax_fit = model_sarimax.fit(disp=False)

predicciones = model_sarimax_fit.forecast(steps=12, exog=exogenas[-12:])  # Usa las variables exógenas

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(serie, label='Observado')
plt.plot(predicciones, label='Predicción', color='red')
plt.legend()
plt.show()
```
